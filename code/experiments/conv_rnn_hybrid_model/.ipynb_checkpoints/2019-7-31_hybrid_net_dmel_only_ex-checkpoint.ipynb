```python
class HybridNet(nn.Module):
    
    def __init__(self, pesudo_input, num_filters, filter_size, rnn_size, fc_out1, fc_out2, dp1, dp2, dp3, 
                 num_rnn_layers=1, rnn_dropout=0):
        super(HybridNet, self).__init__()
        self.conv1 = nn.Conv1d(in_channels=4, out_channels=num_filters, kernel_size=filter_size)
        out = self.conv1(pesudo_input)
        self.layernorm = nn.LayerNorm(out.shape[1:])
        out = nn.MaxPool1d(kernel_size=15, stride=15)(out)
        ####################################################
        print('shape after conv1d {}'.format(out.shape))
        N, C, T = out.shape
        out = torch.transpose(out, 1, 2)
        print('shape before lstm {}'.format(out.shape))
        ####################################################
        # or input_size*seq_len
        self.bi_lstm = nn.LSTM(input_size=C, hidden_size=rnn_size, num_layers=num_rnn_layers,
                              batch_first=True, dropout=rnn_dropout, bidirectional=True)
        out, _ = self.bi_lstm(out)
        print('shape after lstm {}'.format(out.shape))
        N, T, C = out.shape
        #out = torch.transpose(out, 1, 2)
        out = out.reshape(N, -1)
        print('shape after flattening {}'.format(out.shape))
        self.fc1 = nn.Linear(T*C, fc_out1, bias=True)
        self.fc2 = nn.Linear(fc_out1, fc_out2, bias=True)
        self.fc3 = nn.Linear(fc_out2, 1)
        self.p1 = dp1
        self.p2 = dp2
        self.p3 = dp3
        
    def forward(self, seq):
        self.activation_seq = F.relu(self.conv1(seq))
        out = self.layernorm(self.activation_seq)
        out = nn.MaxPool1d(kernel_size=15, stride=15)(out)
        
        out = nn.Dropout(p=self.p1)(out)
        
        #################################################################################
        # Input of LSTM layer should have shape (sequence_length, batch_size, input_size)
        #     - Sequence length here should be the length of activation after downsampling
        #     - Input size should be the number of filters
        #################################################################################
        N, C, T = out.shape
#         out = out.view(bs, 1, -1)
        out = torch.transpose(out, 1, 2)
        out, _ = self.bi_lstm(out)
        out = F.relu(out)
        
        #################################################################################
        # Need to flatten the sequence before feeding them into fully connected layer
        #################################################################################
        N, T, C = out.shape
        #out = torch.transpose(out, 1, 2)
        out = out.reshape(N, -1)
        out = self.fc1(out)
        out = F.relu(out)
        out = nn.Dropout(p=self.p2)(out)
        out = self.fc2(out)
        out = F.relu(out)
        out = nn.Dropout(p=self.p3)(out)
        out = self.fc3(out)
        out = torch.squeeze(out)
        return nn.Sigmoid()(out)
```

```python
class HybridNet(nn.Module):
    
    def __init__(self, pesudo_input, num_filters, filter_sizes, max_pools, conv_dps,
                 rnn_size, 
                 fc_sizes, fc_dps, 
                 num_rnn_layers=1, rnn_dropout=0):
        super(HybridNet, self).__init__()
        self.max_pools = max_pools
        self.convs = []
        self.fcs = []
        self.fc_dps = fc_dps
        self.convs.append(nn.Conv1d(in_channels=4, out_channels=num_filters[0], kernel_size=filter_sizes[0]))
        for i in range(1, len(num_filters)):
            self.convs.append(nn.Conv1d(in_channels=num_filters[i-1], 
                                       out_channels=num_filters[i], 
                                       kernel_size=filter_sizes[i]).cuda())
        out = pesudo_input
        for j, conv_layer in enumerate(self.convs):
            out = conv_layer(out)
            out = nn.MaxPool1d(kernel_size=max_pools[j][0], stride=max_pools[j][1])(out)
        ####################################################
        print('shape after conv1d {}'.format(out.shape))
        N, C, T = out.shape
        out = torch.transpose(out, 1, 2)
        print('shape before lstm {}'.format(out.shape))
        ####################################################
        # or input_size*seq_len
        self.bi_lstm = nn.LSTM(input_size=C, hidden_size=rnn_size, num_layers=num_rnn_layers,
                              batch_first=True, dropout=rnn_dropout, bidirectional=True)
        out, _ = self.bi_lstm(out)
        print('shape after lstm {}'.format(out.shape))
        N, T, C = out.shape
        #out = torch.transpose(out, 1, 2)
        out = out.reshape(N, -1)
        print('shape after flattening {}'.format(out.shape))
        self.fcs.append(nn.Linear(T*C, fc_sizes[0], bias=True))
        for z in range(1, len(fc_sizes)):
            self.fcs.append(nn.Linear(fc_sizes[z-1], fc_sizes[z], bias=True).cuda())
        self.fc_out = nn.Linear(fc_sizes[-1], 1)
        
    def forward(self, seq):
        self.activation_seq = F.relu(self.convs[0](seq))
        out = nn.Dropout(p=conv_dps[0])
        out = nn.MaxPool1d(kernel_size=max_pools[0][0], stride=max_pools[0][1])(self.activation_seq)
        for j, conv_layer in enumerate(conv):
            out = F.relu(conv_layer(out))
            out = nn.Dropout(p=self.conv_dps[j])
            out = nn.MaxPool1d(kernel_size=max_pools[j][0], stride=max_pools[j][1])(out)
        
        #################################################################################
        # Input of LSTM layer should have shape (sequence_length, batch_size, input_size)
        #     - Sequence length here should be the length of activation after downsampling
        #     - Input size should be the number of filters
        #################################################################################
        N, C, T = out.shape
#         out = out.view(bs, 1, -1)
        out = torch.transpose(out, 1, 2)
        out, _ = self.bi_lstm(out)
        
        #################################################################################
        # Need to flatten the sequence before feeding them into fully connected layer
        #################################################################################
        N, T, C = out.shape
        #out = torch.transpose(out, 1, 2)
        out = out.reshape(N, -1)
        for i, fc in enumerate(self.fcs):
            out = F.relu(fc(out))
            out = nn.Dropout(p=self.fc_dps[i])
        out = self.fc_out(out)
        return nn.Sigmoid()(out)
```
