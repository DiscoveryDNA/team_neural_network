This notebook verifies that the data in two folders is the same, in order to test reproducibility of results.
dir1 and dir2 are path variables of the two directories to be compared, and must end with a forward slash.

```python
import os
import hashlib
dir1 = "/home/ubuntu/data/team_neural_network/data/new_output/"
dir2 = "/home/ubuntu/data2/complete_output/"
files1 = os.listdir(dir1)
files2 = os.listdir(dir2)
count = 0
for f in files1:
    hasher1 = hashlib.md5()
    hasher2 = hashlib.md5()
    if not f in files2:
        print("There is a file missing from the second directory")
    else:
        fd1 = open(dir1+f, "rb")
        fd2 = open(dir2+f, "rb")
        buf1 = fd1.read()
        hasher1.update(buf1)
        buf2 = fd2.read()
        hasher2.update(buf2)
        if (hasher1.hexdigest() == hasher2.hexdigest()):
            print(f + " is correct.")
        else:
            print(f + " is incorrect")
            count += 1
            break
print(str(count) + " files incorrect")
if len(files1) == len(files2):
    print("Same number of files")
else:
    print("Different number of files")
```
