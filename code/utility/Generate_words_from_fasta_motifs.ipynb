**Purpose**: This notebook converts motifs from a FASTA file into a standardized word list consistent with Generate_words_by_motif.ipynb 

**Usage**:

- Update the motif_list if necessary
- Modify the motif_path if necessary
- Set corresponding motif_index
- Run the rest of the notebook, and check if the intermediate outputs match expectations
- Finally, set the buffer_file_path, where the list of words is buffered to


```python
import pickle
```

```python
# May add more motifs in the future to this list
motif_list = ['motif_1_fasta.txt']

# Path to where the motif files are stored
motif_path = 'C:/Users/Lanes/Downloads/'
```

```python
# Set the Motif index to the desired number
motif_index = 0

#open and store the data from the desired FASTA file
data = open(motif_path + motif_list[motif_index], 'r').readlines()

data
```

```python
# Iterate through the words in data and store the actual sequence data, discarding the markup language
words = []
for i in range(0,len(data)):
    if i % 2 == 1:
        words.append(data[i][0:-1])
        
words
    
```

```python
buffer_file_path = motif_path + "words_generated_by_" + motif_list[motif_index]

with open(buffer_file_path, "wb") as buff:
    pickle.dump(words, buff)
```
